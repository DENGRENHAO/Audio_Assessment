```python
class_df = pd.read_excel(os.path.join(basepath, 'linebot_data_sound_cate3.xlsx'))
class_df = class_df[['AudioFile_', 'new_cate_3']]
class_df.rename({'AudioFile_': 'AudioFile_name'}, axis=1, inplace=True)
class_df.rename({'new_cate_3': 'Audio_class_cat'}, axis=1, inplace=True)
print(len(class_df))
class_df.head(10)
```

```python
def one_hot_encode(df, column_name):
    encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
    df_encoded = pd.DataFrame(encoder.fit_transform(df[[column_name]]))
    df_encoded.columns = encoder.get_feature_names_out([column_name])
    df.drop([column_name] ,axis=1, inplace=True)
    df = pd.concat([df, df_encoded], axis=1)
    return df
```

```python
class_df = one_hot_encode(class_df, 'Audio_class_cat')
class_df = class_df.dropna()
print(len(class_df))
class_df.head()
```

```python
print(len(class_df))
class_df = class_df.groupby('AudioFile_name').max().reset_index()
print(len(class_df))
class_df.head()
```

### Features preprocessing description
- total 57 columns (including AudioFile_name, Feel_score, Feel_health, and features):
    - changed features:
        - Audio_class: one hot encoded to 10 columns -> 57 to (57 - 1 + 10) = 66 columns
        - Feel_score: change value [1, 2, 3] to 0 and value [4, 5] to 1
        - 時: rename to 'Time_cat', do binning, and one hot to 3 columns. -> 66 to (66 - 1 + 3) = 68 columns
            - night: 23:00-07:00
            - day: 07:00-19:00
            - evening: 19:00-23:00  
        - 周末: rename to is_weekend and change its value
            - if value == 'y': value = 1
            - else: value = 0
        - gender: change its value
            - if value == '男性': value = 1
            - else: value = 0
        - NSS(Noise Sensitivity Scale): change its value
            - if value < 3: value = 0
            - else: value = 1
        - Max_LP: one hot encoded to 10 columns -> 68 to (68 - 1 + 9) = 76 columns (8: 礦鹽用地 並沒有出現於資料中)
        - d_type: one hot encoded to 10 columns -> 76 to (76 - 1 + 3) = 78 columns
        - site_location2: one hot encoded to 10 columns -> 78 to (78 - 1 + 10) = 87 columns
        - ses_level: one hot encoded to 10 columns -> 87 to (87 - 1 + 5) = 91 columns

```python
df = pd.read_excel(os.path.join(basepath, 'linebot_data20221212.xlsx'))
# 'Audio_class' will be merge from linebot_data_class.csv
df = df[['AudioFile_name', 'DBA', 'Feel_score', 'Feel_health', '時', '周末',
         'Min', 'LA90', 'LA50', 'LA10', 'Max', 'Leq', 'LA10_LA90', 'MAX_MIN',
         '大於75dB', 'F65_75dB', 'F55_65dB', 'F45_55dB', '小於45dB',
         '大於4kHz', 'F200_4kHz', '小於200Hz', '100Hz', '125Hz', '160Hz', '200Hz', '250Hz', '315Hz',
         '400Hz', '500Hz', '630Hz', '800Hz', '1000Hz', '1250Hz', '1600Hz', '2000Hz', '2500Hz', '3150Hz',
         '4000Hz', '5000Hz', '6300Hz', '8000Hz', 'age', 'gender', 'BSRS5', 'Noise_exposure',
         'Open Mindedness_BFI', 'Conscientiousness_BFI', 'Extraversion_BFI', 'Agreeableness_BFI',
         'Negative Emotionality_BFI', 'NSS(Noise Sensitivity Scale)', 'Max_LP', 'd_type', 'site_location2', 'ses_level']]
df = df.dropna()

print(len(df))
df = df[abs(df['Feel_score']-df['Feel_health'])<3]
print(len(df))

df = df.loc[df['Feel_score'] != 3]

df['Feel_score'] = np.where((df['Feel_score']==1) | (df['Feel_score']==2), 0, df['Feel_score'])
df['Feel_score'] = np.where((df['Feel_score']==4) | (df['Feel_score']==5), 1, df['Feel_score'])

# df['Feel_score'] = np.where((df['Feel_score']==1) | (df['Feel_score']==2) | (df['Feel_score']==3), 0, df['Feel_score'])
# df['Feel_score'] = np.where((df['Feel_score']==4) | (df['Feel_score']==5), 1, df['Feel_score'])

# df['Feel_health'] = np.where((df['Feel_health']==1) | (df['Feel_health']==2) | (df['Feel_health']==3), 0, df['Feel_health'])
# df['Feel_health'] = np.where((df['Feel_health']==4) | (df['Feel_health']==5), 1, df['Feel_health'])

df['NSS(Noise Sensitivity Scale)'] = np.where((df['NSS(Noise Sensitivity Scale)'] < 3), 0, df['NSS(Noise Sensitivity Scale)'])
df['NSS(Noise Sensitivity Scale)'] = np.where((df['NSS(Noise Sensitivity Scale)']>=3), 1, df['NSS(Noise Sensitivity Scale)'])

df.rename({'周末': 'is_weekend'}, axis=1, inplace=True)
df['is_weekend'] = np.where(df['is_weekend']!='y', 0, df['is_weekend'])
df['is_weekend'] = np.where(df['is_weekend']=='y', 1, df['is_weekend'])

df['gender'] = np.where(df['gender']!='男性', 0, df['gender'])
df['gender'] = np.where(df['gender']=='男性', 1, df['gender'])

df = pd.merge(df, class_df, on="AudioFile_name", how="inner")

df.rename({'時': 'Time_cat'}, axis=1, inplace=True)
# df = df.assign(Time_cat=pd.cut(df['Time_cat'], 
#                                bins=[0, 7, 17, 20, 24], 
#                                labels=['1', '2', '3', '4']))

df = df.assign(Time_cat=pd.cut(df['Time_cat'], 
                               bins=[-1, 6, 18, 22, 24], 
                               labels=['night1', 'day', 'evening', 'night2']))

df = one_hot_encode(df, 'Time_cat')
df['Time_cat_night'] = df[['Time_cat_night1', 'Time_cat_night2']].max(axis=1)
df = df.drop(['Time_cat_night1', 'Time_cat_night2'], axis=1)

df = one_hot_encode(df, 'Max_LP')
df = one_hot_encode(df, 'd_type')
df = one_hot_encode(df, 'site_location2')
df = one_hot_encode(df, 'ses_level')

df = df.dropna().sample(frac=1, random_state=pd_random_seed).reset_index(drop=True)
print(df.shape)
df.head(10)
```

```python
def next_record(df):
    for index, row in df.iterrows():
        filename = row['AudioFile_name']
        feel_score = row['Feel_score']
        feel_health = row['Feel_health']
        feature = [row[1]]+list(row[4:])
        yield feel_score, feel_health, filename, feature
```

```python
# for catboost
feel_scores = []
feel_healths = []
filenames = []

features = []

print(len(df))
i = 0
data_generator = next_record(df)
for feel_score, feel_health, filename, feature in data_generator:
    feel_scores.append(feel_score)
    feel_healths.append(feel_health)
    filenames.append(filename)
    features.append(feature)
    i += 1
    if i%500==0:
        print(f"dataset prepared: {i}")
print(len(features[0]))
```

```python
# for catboost
filenames = np.asarray(filenames, dtype='object')
feel_scores = np.asarray(feel_scores, dtype='object')
feel_healths = np.asarray(feel_healths, dtype='object')

features = np.asarray(features, dtype='object')
```
